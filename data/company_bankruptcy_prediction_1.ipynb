```python
df1 = df.copy()
```

```python
y = df1['Bankrupt?']
X = df1.drop(columns = 'Bankrupt?',axis=1)
print(y.shape,X.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,train_size=0.8,random_state=678)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
scaler = MinMaxScaler()
#scaler = StandardScaler()


X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.fit_transform(X_test)
print(type(X_train_sc))
print(type(X_test_sc))
```

```python
def evaluate_classifiers(X_train_sc, X_test_sc, y_train, y_test):

  # Define models and their names
  models = {
      "Logistic Regression": LogisticRegression(max_iter=1000),
      "Support Vector Classifier": SVC(),
      "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),
      "Decision Tree": DecisionTreeClassifier(),
      "Random Forest": RandomForestClassifier(n_estimators=100), 
      "MLPClassifier": MLPClassifier(),
      "RandomForestClassifier": RandomForestClassifier(), 
      "GradientBoostingClassifier": GradientBoostingClassifier()
  }

  # Train models and store scores
  results = []
  for name, model in models.items():
    model.fit(X_train_sc, y_train)
    y_pred = model.predict(X_test_sc)
    score = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test,y_pred)

    results.append({"Model": name, "Accuracy_Score": score,"Precision": precision})
    
  # Create and return DataFrame
  df_results = pd.DataFrame(results).sort_values(by=['Accuracy_Score',"Precision"],ascending=False, ignore_index= True)
  return df_results

# Example usage (assuming you have your training and testing data)
df_scores = evaluate_classifiers(X_train_sc, X_test_sc, y_train, y_test)
print(df_scores)
```

```python
print(f"The best model is {df_scores.iloc[0,0]} with the Accuracy score of {df_scores.iloc[0,1]}.")
```
